```python
import pandas as pd
```

```python
# Charger le fichier Excel dans un DataFrame
data = pd.read_excel("Filiere_Etudiant.xlsx")

# Charger le deuxième fichier Excel dans un DataFrame
data_nom_prenom = pd.read_excel("Groupes_S2_V1.xlsx")

# Fusionner les deux DataFrames en utilisant la colonne 'Nom' comme clé de fusion
data_merged = pd.merge(data, data_nom_prenom, on='Nom')


data_merged.head()
```

```python
import pandas as pd

# Supprimer la partie après le premier tiret dans la colonne de la filière
data_merged['Classe'] = data_merged['Classe'].apply(lambda x: x.split('-', 1)[0] if '-' in x else x)

# Supprimer les espaces éventuels avant et après
data_merged['Classe'] = data_merged['Classe'].str.strip()

# Supprimer le préfixe 'E4' éventuel
data_merged['Classe'] = data_merged['Classe'].str.replace('E4', '')

data_merged.head()
```

```python
data_clear = data_merged[['Classe','Nom','Prénom_x','Affectation créneau E','Affectation créneau F']].copy()
```

```python
data_clear = data_clear.rename(columns={'Prénom_x': 'Prénom'})

data_clear.head()
```

# Analyse Statistiques

```python
import matplotlib.pyplot as plt
```

```python
def pie_chart(data):
    
    # Créer un diagramme circulaire
    plt.figure(figsize=(8, 8))
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
    plt.title("Répartition des élèves en fonction de leur affectation dans le créneau E")
    plt.axis('equal')  # Assure que le diagramme est un cercle et non pas une ellipse
    # Ajouter la légende  
    plt.show()
```

```python
data_clear['Classe'] = data_clear['Classe'].str.strip()

data_clear['Affectation créneau E'].unique().tolist()
```

## Filiere AIC

```python
data_AIC = data_clear[data_clear['Classe'] == 'AIC']['Affectation créneau E'].value_counts('')
pie_chart(data_AIC)
```

## Filiere DSIA

```python
data_DSIA = data_clear[data_clear['Classe'] == 'DSIA']['Affectation créneau E'].value_counts('')

pie_chart(data_DSIA)

data_DSIA = data_clear[data_clear['Classe'] == 'DSIA']['Affectation créneau F'].value_counts('')

pie_chart(data_DSIA)
```
